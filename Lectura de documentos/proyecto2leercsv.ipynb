```python
import pandas as pd
```

```python
df_csv = pd.read_csv("exported_db_data.csv",index_col = 0)
df_csv.head()
```

```python
df_csv.shape
```

```python
def propiedades (df):
    a = ("El número de filas y columnas es: ", df.shape) 
    b = ("Las columnas son: ", df.columns)
    c = ("El tipo de datos es: ", df.dtypes)
    d = ("Cantidad de nulos: ",df.isna().sum()) 
    e = ("Descripción general: ", df.describe(include = "object").T)
    f = ("Las primeras filas son: ", df.head())
    
    print(a, "\n", b, "\n", c, "\n", d, "\n", e,  "\n", f,)
    
propiedades(df_csv)
```

```python
minus = {col :  col.lower() for col in df_csv.columns}
df_csv.rename(columns = minus, inplace= True)
```
